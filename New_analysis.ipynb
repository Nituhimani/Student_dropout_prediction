```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("dataset.csv")
df.head()
```

```python
df1=df.drop(df[df['Target']=='Enrolled'].index)
df1.head()
```

```python
df.nunique()
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtype=='int64':
            print(f'{column}:{df[column].unique}')
      
                  
```

```python
print_unique_col_values(df1)
```

```python
pd.crosstab(df['Marital status'],df1.Target).plot(kind='bar')
```

```python
selected_colums=['Marital status','Course','Daytime/evening attendance','Previous qualification','Mother\'s qualification','Father\'s qualification','Mother\'s occupation','Father\'s occupation','Displaced','Scholarship holder','Age at enrollment','Curricular units 1st sem (evaluations)','Curricular units 2nd sem (evaluations)','GDP','Gender','Tuition fees up to date','Educational special needs','Target']
df2=df1[selected_colums]
df2.head(2)
```

```python
df2.nunique()
```

```python
cols_to_scale=['Marital status','Course','Previous qualification','Mother\'s occupation','Mother\'s qualification','Father\'s occupation','Father\'s qualification','Age at enrollment','Curricular units 1st sem (evaluations)','Curricular units 2nd sem (evaluations)','GDP']
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
df2[cols_to_scale]=scalar.fit_transform(df2[cols_to_scale])
```

```python
df2.head()
```

```python
df2['Dropout']=df['Target'].apply(lambda x:1 if x=='Dropout' else 0)
```

```python
df2.head(2)
```

```python
df3=df2.drop('Target',axis=1)
df3.head()
```

```python
#feature selection
corr_matrix=df3.corr(method="pearson")
corr_matrix['Dropout']
```

```python
X=df3.drop('Dropout',axis=1)
y=df3['Dropout']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score

score_lr=cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'),X,y,cv=3)
score_svc=cross_val_score(SVC(gamma='auto'),X,y,cv=3)
score_rf=cross_val_score(RandomForestClassifier(n_estimators=30),X,y,cv=3)
score_dt=cross_val_score(DecisionTreeClassifier(),X,y,cv=3)
```

```python
import numpy as np
np.average(score_lr)
```

```python
np.average(score_svc)
```

```python
np.average(score_dt)
```

```python
np.average(score_rf)
```

```python
#ratio of dropout have scholarship and no scholarship
df3=df2.drop(df2[df2['Dropout']==0].index)
plt.figure(figsize=(5,5))
plt.title("Scholarship holds effecting dropout")
plt.pie(df3['Scholarship holder'].value_counts(),labels=['No_scholarship','Have_scholarship'],autopct='%1.2f%%',shadow=True)
```

```python
plt.figure(figsize=(5,5))
plt.title("Educational special needs effecting dropout")
plt.pie(df3['Educational special needs'].value_counts(),labels=['Dont_have_Educational special needs','Have_Educational special needs'],autopct='%1.2f%%',shadow=True)
```

```python
plt.figure(figsize=(5,5))
plt.title("Attendance effecting dropout")
plt.pie(df3['Daytime/evening attendance'].value_counts(),labels=['Dont_have_attendance','Have_attendance'],autopct='%1.2f%%',shadow=True)
```

```python
plt.figure(figsize=(5,5))
plt.title("Tuition fees effecting dropout ")
plt.pie(df3['Tuition fees up to date'].value_counts(),labels=['Dont_have_fees_upto_date','Have_fees_upto_date'],autopct='%1.2f%%',shadow=True)
```

```python
plt.figure(figsize=(5,5))
plt.title("Displaced effecting dropout ")
plt.pie(df3['Displaced'].value_counts(),labels=['Not_Displaced','Have_Displaced'],autopct='%1.2f%%',shadow=True)
```
