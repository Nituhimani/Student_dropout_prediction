```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("dataset.csv")
df.head()
```

```python
df1=df.drop(df[df['Target']=='Enrolled'].index)
df1.head()
```

```python
df.nunique()
```

```python
df['Marital status'].unique()
```

```python
pd.crosstab(df['Marital status'],df1.Target).plot(kind='bar')
```

```python
selected_colums=['Marital status','Course','Daytime/evening attendance','Previous qualification','Mother\'s qualification','Father\'s qualification','Mother\'s occupation','Father\'s occupation','Displaced','Scholarship holder','Age at enrollment','Curricular units 1st sem (evaluations)','Curricular units 2nd sem (evaluations)','GDP','Gender','Tuition fees up to date','Educational special needs','Target']
df2=df1[selected_colums]
df2.head(2)
```

```python
df2.nunique()
```

```python
cols_to_scale=['Marital status','Course','Previous qualification','Mother\'s occupation','Mother\'s qualification','Father\'s occupation','Father\'s qualification','Age at enrollment','Curricular units 1st sem (evaluations)','Curricular units 2nd sem (evaluations)','GDP']
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
df2[cols_to_scale]=scalar.fit_transform(df2[cols_to_scale])
```

```python
df2.head()
```

```python
df2['Dropout']=df['Target'].apply(lambda x:1 if x=='Dropout' else 0)
```

```python
df2.head(2)
```

```python
df3=df2.drop('Target',axis=1)
df3.head()
```

```python
#feature selection
corr_matrix=df3.corr(method="pearson")
corr_matrix['Dropout']
```

```python
sns.heatmap(corr_matrix)
```

```python
X=df3.drop('Dropout',axis=1)
y=df3['Dropout']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model_lr=LogisticRegression()
model_lr.fit(x_train,y_train)
```

```python
len(model_lr.coef_[0]),len(X.columns)
```

```python
model_lr.score(x_test,y_test)
```

```python
y_pred_lr=model_lr.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_lr)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted_lr')
plt.ylabel('Truth')
```

```python
model_svc=SVC()
model_svc.fit(x_train,y_train)
model_svc.score(x_test,y_test)
```

```python
y_pred_svc=model_svc.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_svc)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted_svc')
plt.ylabel('Truth')
```

```python
model_rf=RandomForestClassifier()
model_rf.fit(x_train,y_train)
model_rf.score(x_test,y_test)
```

```python
y_pred_rf=model_rf.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_rf)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted_rf')
plt.ylabel('Truth')
```

```python
model_dt=DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
model_dt.score(x_test,y_test)
```

```python
y_pred_dt=model_dt.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_dt)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted_dt')
plt.ylabel('Truth')
```

```python
result=pd.DataFrame()
result['column']=X.columns
result['coefficient']= model_lr.coef_[0]
result
```

```python
X.corrwith(y).plot.bar(figsize=(16,4),rot=90,grid=False)
plt.title('correlation with dropout',
            fontsize=25,color='Blue')
plt.show()

```

```python
df_drop = df2[df2['Target']=='Dropout']
df_selected = df_drop[['Scholarship holder','Educational special needs','Daytime/evening attendance','Tuition fees up to date']]
col_name = [ ]
zero= [ ]
one = [ ]
for i in df_selected:
    print(i)
    z,o = df_selected[i].value_counts()
    col_name.append(i)
    zero.append(z)
    one.append(o)
output = pd.DataFrame()
output['column_name'] = col_name
output['zero_count'] = zero
output['one_count'] = one
output['Total'] = output['one_count']+output['zero_count']

```

```python
output.to_csv('dropout.xlsx')
```
